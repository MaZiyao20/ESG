Calculate ESG_Score percentile rank and ESG uncertainty

```python
#open csv document downloaded from Github
ESG = pd.read_csv(r"D:\PhD\ESG Lecture\ESG.csv")
ESG
```

```python
#Calculate every year's ESG Score percentile rank
grouped = ESG.groupby(['Date']) #tell python our calculation is based on ['Date'] (based on every year)

#Calculate ESG score's percentile rank
ESG['Sustainalytics_Rank'] = grouped['Sustainalytics_ESG_Score'].transform(lambda x: x.rank(pct=True)) 
ESG['Refinitive_Rank'] = grouped['Refinitiv_ESG_Score'].transform(lambda x: x.rank(pct=True))
ESG['MSCI_KLD_Rank'] = grouped['MSCI_KLD_Score'].transform(lambda x: x.rank(pct=True))
ESG
```

```python
#Calculate ESG Uncertainty (Standard deviation of three different ESG scores) 
ESG['Uncertainty'] = ESG[["Sustainalytics_Rank", "Refinitive_Rank", "MSCI_KLD_Rank"]].std(axis=1, ddof=1)
ESG
```

```python
#Calculate average ESG score using ESG percentile rank
ESG['Total_ESG'] = (ESG['Sustainalytics_Rank']+ESG['Refinitive_Rank']+ESG['MSCI_KLD_Rank'])/3
ESG
```

```python
#Calculate average ESG score's percentile rank
grouped = ESG.groupby(['Date'])
ESG['Total_ESG_Rank'] = grouped['Total_ESG'].transform(lambda x: x.rank(pct=True))
ESG
```

Construct and Sort portfolio based on quintile

```python
#Construct low ESG rating portfolio based on quintile
portfolio1 = ESG[ESG['Total_ESG_Rank']<=0.2]
portfolio1
```

```python
#Use portfolio1(low ESG rating portfolio) to sort ESG uncertainty's percentile rank 
grouped = portfolio1.groupby(['Date'])
portfolio1['Uncertainty_rank'] = grouped['Uncertainty'].transform(lambda x: x.rank(pct=True))
portfolio1
```

```python
#Use portfolio1 (low ESG rating portfolio) to construct low ESG rating and low ESG uncertainty portfolio 
portfolio1_1 = portfolio1[portfolio1['Uncertainty_rank']<=0.2]
portfolio1_1
```

Calculate market-value weighted stock return

```python
#Calculate every month's total market value
result = portfolio1_1.groupby(['Year-Month'],as_index=False)[['Market_Value']].sum()
result
```

```python
#Rename variable to prepare to merge every month's total market value with low ESG rating and low ESG uncertainty portfolio
result = result.rename(columns = {'Market_Value':'Sum'})
result
```

```python
#Merge market-value weighted stock return and portfolio with low ESG rating and low ESG uncertainty 
portfolio1_1 = pd.merge(result,portfolio1_1,how='right',on='Year-Month')
portfolio1_1
```

```python
#Calculate ervery month's firm market-value weighted stock return
portfolio1_1['return'] = portfolio1_1['Stock_Return']*portfolio1_1['Market_Value']/portfolio1_1['Sum']
portfolio1_1
```

```python
#Calculate every month's portfolio return
portfolio1_1 = portfolio1_1.groupby(['Year-Month'],as_index=False)[['return']].sum()
portfolio1_1
```
